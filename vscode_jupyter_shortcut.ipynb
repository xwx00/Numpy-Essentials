# VSCode Jupyter Notebook 快捷键总结

## 🧠 模式说明
- **命令模式（Command Mode）**：按 `Esc` 进入（操作 cell）
- **编辑模式（Edit Mode）**：按 `Enter` 进入（编辑内容）

---

## 📦 创建 / 插入 Cell

| 操作 | 快捷键 |
|------|--------|
| 在下方插入 cell | `B` |
| 在上方插入 cell | `A` |
| 在当前 cell 下插入（并进入编辑） | `Alt + Enter` |
| 运行当前 cell 并在下方创建 | `Shift + Enter` |

---

## 🔄 切换 Cell 类型

| 操作 | 快捷键 |
|------|--------|
| 转为 Code Cell | `Y` |
| 转为 Markdown Cell | `M` |

---

## 🧭 Cell 之间跳转

| 操作 | 快捷键 |
|------|--------|
| 选中上一个 cell | `K` 或 `↑` |
| 选中下一个 cell | `J` 或 `↓` |

---

## ✂️ Cell 操作

| 操作 | 快捷键 |
|------|--------|
| 删除 cell | `D` + `D` |
| 复制 cell | `C` |
| 剪切 cell | `X` |
| 粘贴到下方 | `V` |
| 粘贴到上方 | `Shift + V` |

---

## ▶️ 运行 Cell

| 操作 | 快捷键 |
|------|--------|
| 运行当前 cell | `Ctrl + Enter` |
| 运行并跳到下一个 | `Shift + Enter` |
| 运行并新建 cell | `Alt + Enter` |

---

## 🔼🔽 移动 Cell

| 操作 | 快捷键 |
|------|--------|
| 上移 cell | `Alt + ↑` |
| 下移 cell | `Alt + ↓` |

---

## ✏️ 编辑相关

| 操作 | 快捷键 |
|------|--------|
| 进入编辑模式 | `Enter` |
| 退出编辑模式 | `Esc` |

---

## 🔍 其他常用

| 操作 | 快捷键 |
|------|--------|
| 显示所有快捷键 | `H` |
| 切换行号 | `L` |
| 合并 cell | `Shift + M` |

---

# Universal functions(ufuncs)

```python
x = np.arange(5,10)
np.square(x)
print(x)
y = np.ones(5) * 10
print(np.mod(y,x))
print(y)
np.minimum(x,7)
```

```python
# more advanced ufuncs
z = np.repeat(x,3).reshape(-1,3)
print(z)
```

```python
np.median(z)
```

```python
np.median(z, axis = 0)
```

```python
np.median(z, axis = 1)
```

# Broadcasting and shape manipulation

Two conditions for broadcasting:
1. Two arrays should be of equal dimensions
2. One of them is 1

```python
x = np.array([0,10,20]).repeat(3).reshape(-1,3)
print(x)
print(x.shape)

y = np.array([1,2,3])
print(y.shape)
print(x+y)
print(x+y[:, None])
print(x+y[None,:])
```

```python
x = np.array([[0],[10],[20]])
print(x)
print(x+y)
```

```python
x = np.arange(1000000)
x.shape = 100, 100, 100
%timeit x.flatten() # create a new copy of x
%timeit x.ravel() # change the view of x
```

```python
x = np.arange(0, 10, 2)
y = np.arange(0,-5,-1)
print(np.vstack([x,y]))
print(np.hstack([x,y]))
```

# A boolean mask

```python
x = np.array([1,3,-1,5,7,-1])
mask = (x<0)
mask
```

```python
x[mask] = 0
x
```

```python
x = np.random.random(50)
(x > .5).sum() # count the number of numbers greater than 0.5 in the random array
```
